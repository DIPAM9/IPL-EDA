```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
delivery=pd.read_csv('deliveries.csv')
delivery.head()
```

```python
mask=delivery[delivery['bowler']=='R Ashwin']
d={'CH Gayle'}
y=mask[mask['batsman'].isin(d)]
newdf=y[y['batsman_runs']==6.0]
newdf.groupby('bowler')['batsman'].count()
```

```python
def bowler_hit_six(player):
    mask=delivery[delivery['bowler']==player]
    d={'CH Gayle'}
    y=mask[mask['batsman'].isin(d)]
    newdf=y[y['batsman_runs']==6.0]
    z=newdf.groupby('bowler')['batsman'].count()
    return z
```

```python
bowler_hit_six('R Ashwin')
```
